__[4] clean the data__

Make all dataframes contain the same countries and years

```python
def countries():
    """
    Returns the set of countries that apprear in all dataframes.
    -------
    countries_v : SET
    The set of countries that appear in all dataframes.

    """
    population_lst = population['country'].tolist()
    life_expec_lst = life_expec['country'].tolist()
    income_per_person_lst = income_per_person['country'].tolist()
    child_per_woman_lst = child_per_woman['country'].tolist()
    child_mortality_lst = child_mortality['country'].tolist()
    co2_per_person_lst = co2_per_person['country'].tolist()
    
    countries_v = set(population_lst).intersection(life_expec_lst)
    countries_v = countries_v.intersection(income_per_person_lst)
    countries_v = countries_v.intersection(child_per_woman_lst)
    countries_v = countries_v.intersection(child_mortality_lst)
    countries_v = countries_v.intersection(co2_per_person_lst)
    
    return countries_v
    
    
    
def country_year(df,name):
    """
    This function is used to make all dataframes have the same countries & years, 
    and drop the rest.
    PARAMETERS:
        df: DATAFRAME
        The dataframe to be modified.
        name: STR
        The choosen title of the dataframe, used in displaying the results.
    RETURNS:
        df: DATAFRAME
        The modified dataframe
    """
    df_mask = df['country'].apply(lambda x: x in countries())
    df = df.loc[df_mask,'country':'2018']
    print(name + ' MOD')
    print(df.info())
    return df
```

```python
population = country_year(population,'POPULATION')
life_expec = country_year(life_expec,'LIFE EXPECTANCY')
income_per_person = country_year(income_per_person,'INCOME PER PERSON')
child_per_woman = country_year(child_per_woman,'CHILD PER WOMAN')
child_mortality = country_year(child_mortality,'CHILD MORTALITY 0 to 5 years PER 1000 BORN')
co2_per_person = country_year(co2_per_person,'CO2 EMISSIONS TONNES PER PERSON')
```

Transform population, income_per_person & CO2 emissions data into numeric values

```python
def sym2num(x):
    """
    Transform the number from this format "NO_suffix" (such as "3.25M")
    to its equivalent numerical value (such as 3250000.0).
    where,
    'M' stands for millions
    'k' stands for thousands
    'B' stands for Billions
    '' stands for micros

    Parameters
    ----------
    x : STR
        The value as a string.

    Returns
    -------
    x: FLOAT
        The value as a float

    """
    if type(x) == str:
        if x[-1] == 'M':
            return float(x[:-1])*(10**6) 
        elif x[-1] == 'k':
            return float(x[:-1])*(10**3)
        elif x[-1] == 'B':
            return float(x[:-1])*(10**9)
        elif x[-1] == '':
            return float(x[:-1])*(10**(-6))
        else:
            return float(x)
    else:
        return x

def transform2num(df,name):
    countries_p = df['country']
    df = df.loc[:,'1800':'2018'].applymap(sym2num)
    df.insert(0,'country',countries_p)
    print(name + ' the first 5 rows\n')
    print(df.head())
    return df
 
```

_For all countries, Is there a main trend for each metric across time?_

```python
def plt_all(df,name):
    rows_no = df.shape[0]
    years = df.columns
    years = list(map(lambda z: float(z), years[1:]))
    for i in range(0,rows_no):
        y = list(df.iloc[i].tolist())
        y = list(map(lambda z: float(z), y[1:]))
        plt.plot(years,y)
    plt.title('World '+name+' History Graph',pad=20)
    plt.xlabel('Year')
    plt.ylabel(name)
    plt.show()
```

_Effect of Income per person on every other metric_

```python
def plt_against(df_x,df_y,name_x,name_y):
    rows_no = df_x.shape[0]
    for i in range(0,rows_no):
        x = list(df_x.iloc[i].tolist())
        x = list(map(lambda z: float(z), x[1:]))
        y = list(df_y.iloc[i].tolist())
        y = list(map(lambda z: float(z), y[1:]))
        plt.scatter(x,y)
    plt.title(name_y+' AGAINST '+name_x,pad=20)
    plt.xlabel(name_x)
    plt.ylabel(name_y)
    plt.show()
```

```python
plt_against(income_per_person,population,'Income Per Person','Population')
```

__However, Countries of population over 4E8 have the lowest income values ever, and all countries with income values over 50000 have small population.__

# Conclusions

### Globally

To summarize .. health condition, income and population are all running in a growth phase __Globally__ .. mainly since 1900.

From a systemic point of view .. advances in technology and science in the 20th century provided the ground for these growth trends.

Life expectancy is a reflection for the overall health conditions .. but we should be somehow skeptic to take it as a solid indicator .. as people's lives can be extended with life aiding medical machines but these machines mostly don't improve the health condition.

By looking to how many children per woman, the trend is decreasing .. using lay man's logic this sounds somehow contradictory with the continuous population growth.

This means that the __CAPACITY OF GROWTH__ limits the individuals' ability to reproduce.

With more people who are able to reproduce, the individual's share is decreasing .. this means the capacity of growth is not growing enough to satisfy even a __constant reproduction share__ over generations.

_Constant reproduction share_ over generations is modeled mathematically with perfect exponential growth function.

So, any given country with population growth that's approx. quadratic (less growth rate than exponential) must suffer from a degredation of the reproduction ability per individual over time.

Child mortality has remarkably enhanced, and the 2018's condition is the best ever globally for almost all countries.

Though, across history, this decrease is combined with small number of increasing spikes in mortality between 1900 and 2000.

Income Per Person also follows the growth trend .. some countries enjoy remarkably higher increase rate than others .. and some countries have small growth rates .. But in short, most countries have constant approx. linear growth.

Some countries enjoyed increasing spikes in income at the time between 1960 and 1990, but they were all hit with a rebound correction, then the approx. linear trend started to work again. 

CO2 emissions is increasing, with the occurance of rebound effect, one time around 1960 and the second time is around 2015.

### Regarding Income Per Person

The minimum possible life expectancy value increases with the increase of income, and this is the beauty of the positive effect between income and life expectancy.
